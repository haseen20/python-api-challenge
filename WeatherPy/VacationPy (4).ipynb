```python
import geoviews as gv
import hvplot.pandas
from geoviews import opts

# Enable geoViews extension
gv.extension('bokeh', 'matplotlib')

# Create a map with points for each city based on latitude and longitude
city_points = city_data_df.hvplot.points(
    'Lng', 'Lat',
    geo=True,
    size='Humidity',
    color='Humidity',
    hover_cols=['City', 'Country', 'Max Temp', 'Humidity'],
    tiles='OSM'
)

# Display the map
city_points
```

```python
# Filter for ideal weather conditions
ideal_weather_df = city_data_df[
    (city_data_df["Max Temp"] < 27) &
    (city_data_df["Max Temp"] > 21) &
    (city_data_df["Wind Speed"] < 4.5) &
    (city_data_df["Cloudiness"] == 0)
]

# Preview the filtered data
ideal_weather_df.head()
```

```python
# Create a map with points for cities matching the ideal weather conditions
ideal_cities_map = ideal_weather_df.hvplot.points(
    'Lng', 'Lat',
    geo=True,
    size='Humidity',
    color='Humidity',
    hover_cols=['City', 'Country', 'Max Temp', 'Humidity'],
    tiles='OSM'
)

# Display the map
ideal_cities_map
```
